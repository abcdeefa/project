```python
import pandas as pd #판다스 불러오기
from sentence_transformers import SentenceTransformer #sentence-transformers 불러오기
from sklearn.metrics.pairwise import cosine_similarity #유사도 계산을 위한 cosine_similarity 불러오기
```

```python
model = SentenceTransformer("jhgan/ko-sroberta-multitask")#모델 불러오기
```

```python
corpus = pd.read_csv("https://raw.githubusercontent.com/songys/Chatbot_data/refs/heads/master/ChatbotData.csv")#데이터셋 집어넣기
corpus["label"] = corpus['Q'].map(lambda x: list(model.encode(x)))#label에 ko-sroberta-multitask에 있는 encode를 사용해 인코딩으로 변환해 넣기
```

```python
print("1.챗봇\n2.종료") #기능안내
while 1: #지속
  a = int(input("원하시는 번호를 입력하세요:")) #번호 저장
  if a==1: #번호에 따른 기능
    text = input("문장을 입력하세요:") # 문장 입력받기
    label = model.encode(text) # 입력받은거 인코딩하기
    corpus['similarity'] = corpus['label'].map(lambda x: cosine_similarity([label],[x]).squeeze()) #label 다 돌면서 유사도 구하기
    answer = corpus.loc[corpus['similarity'].idxmax()]#가장 높은 유사도 행렬번호 구하기
    print(answer['A'])# 그 번호의 답 출력하기
  elif a==2: #번호에 따른 기능
    break
  else: #잘못된 번호를 입력받았을 때 돌아가기
    print("잘못된 입력입니다.")
    continue
```
